# TOP 5 TRENDING VIDEOS IN EACH COUNTRY

```python
def top5(country):
    if country == country:
            temp = data[data['country']==country]
            temp = temp.sort_values(by = 'views',ascending = False)
            temp = temp[["title","video_id","channel_title","category","views","likes"]]
            temp = pd.DataFrame(temp)
    print(temp.head(5))
```

```python
top5("IN")
```
